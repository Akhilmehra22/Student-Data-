# Step 6: Import maplotlib and seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 7: Analysis on Store_Sales based on Items_Available, Daily_Customer_Count, and Store_Area

We will use Seaborn lmplot for that analysis.

What is Seaborn lmplot?

Seaborn's lmplot is a 2D scatterplot with an optional overlaid regression line. This is useful for comparing numeric variables. Logistic regression for binary classification is also supported with lmplot .

```python
analysis_all = ["Items_Available", "Daily_Customer_Count", "Store_Area"]
color_all = ["black", "red", "green"]
title_all = ["Store Sales vs Items Available", "Store Sales vs Daily Customer Count", "Store Sales vs Store Area"]
for i,j,k in zip(analysis_all,color_all, title_all):
    sns.lmplot(x ="Store_Sales" ,y= i, data = df, lowess = True, aspect = 3, line_kws = {'color':j}).set(title=k)
    plt.show()
```

> **Observation:** 
> 1. The Regression line describes that the sale is more when there are more items in the store.
> 2. The Regression line is almost plain when we look at Store Sales based on Daily Customer Count. So we can say that store sales aren't
>    dependent on Daily Customer Count. That's surprising! 
> 3. Store Sales are a little bit more when Store Area number is high as per Regression Line.

# Relation Between Daily_Customer_Count and Items_Available

Here, we will use the seaborn catplot to visualize. 

What is Seaborn Catplot? 

Seaborn catplot shows frequencies (or optionally fractions or percents) of the categories of one, two, or three categorical variables.

```python
df = df.sort_values(by = "Daily_Customer_Count")
sns.catplot(x = "Daily_Customer_Count",y = "Items_Available", data = df, aspect =4)
plt.xticks(rotation =90)
plt.show()
```

# Final Step: Store Sales Visualization

We will seaborn violinplot to visualize Store_Sales Data.

What is Seaborn violinplot?

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.

```python
df = df.sort_values(by = "Daily_Customer_Count")
sns.violinplot(y = "Store_Sales", data = df)
plt.show()
```

> **Observation: Here, we can see the median of Store Sales is near 60000. The interquartile range is more than 40000 and below 80000.**
